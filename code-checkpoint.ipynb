```python
def speak(audio): 
    engine.say(audio) 
    engine.runAndWait() 
```

```python
def wishMe(): 
    # Load the dataset
    df = pd.read_csv("C:/Users/Rahul/OneDrive/Desktop/Major Project/Minor project/datasets/greetings.csv", header=None)  # Replace with your CSV file path
    # Select a random sentence from the 'sentence' column
    random_sentence = df[0].sample().iloc[0]
    print("Random sentence:", random_sentence)
    speak(random_sentence)
```

```python
def takeCommand():
    r = sr.Recognizer() 
    with sr.Microphone(device_index=0) as source: 
        print("Listening...") 
        r.pause_threshold = 0.5 
        audio = r.listen(source) 
    try: 
        print("Recognizing..")
        query = r.recognize_google(audio, language='en-in') 
        print(f"User said:{query}\n")
        speak(query)
    except Exception as e: 
        print("Unable to Recognize your voice.") 
        return "None" 
    return query
```
